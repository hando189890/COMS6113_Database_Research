```python
import jsonlines
import os
train_file = "train.tables.json"
train_path = os.path.join(root_path, train_file)

id=[]
header=[]
types=[]
rows=[]
with jsonlines.open(train_path) as f:
  for line in f.iter():
    id.append(str(line['id']))
    header.append([str(ele) for ele in line['header']])
    types.append([ele.replace("真实的", "实数") if ele == "真实的" else ele for ele in line['types']])
    rows.append([[str(ele) for ele in j] for j in line['rows']])
d={'id':id,'header':header,'types':types,'rows':rows}


train_table_info=pd.DataFrame(d)
train_table_info
```

```python
#merging both tables
merged_data=train_data.merge(train_table_info,on='id')
merged_data
print(merged_data.iloc[10])
print(len(merged_data))
print(merged_data.iloc[10]['sql'])
print(type(merged_data))
```

```python
#Converting dictionary sql answer into sql text f
from tqdm import tqdm 
sql_in_text=[]
aggregate=['','MAX','MIN','COUNT','SUM','AVG']
operator=['=','<','>']
word_operator=['=','<','>']

for i in tqdm(range(len(merged_data))):
    sql=merged_data.sql.iloc[i]
    col=merged_data.header.iloc[i][sql['sel'][0]]
    agg=aggregate[sql['agg'][0]]
    cond_col=[]
    cond_opr=[]
    cond_value=[]
    for j in sql['conds']:
      try:
        cond_col.append(merged_data.header.iloc[i][j[0]])
        cond_opr.append(word_operator[j[1]])
        cond_value.append(j[2])
      except:
        print("\nError Sql: ", sql)
    
    resulted_sql="select "
    if(agg==''):
      resulted_sql+=col+" where "
    else:
      resulted_sql+=agg+"("+col+") where "

    for j in range(len(cond_col)):
      try:
          if(j!=0):
            resulted_sql+="and "
          resulted_sql+=cond_col[j]+' '+cond_opr[j]+' '
          resulted_sql+=str(cond_value[j])+' '
      except:
          print("\nError cond_col index: ", j)

    sql_in_text.append(resulted_sql)
```

**Let's calculate the inputs and targets lengths**

```python
def map_to_length(x):
  x["input_len"] = len(tokenizer(x["input"]).input_ids)
  x["input_longer_256"] = int(x["input_len"] > 256)
  x["input_longer_128"] = int(x["input_len"] > 128)
  x["input_longer_64"] = int(x["input_len"] > 64)
  x["out_len"] = len(tokenizer(x["target"]).input_ids)
  x["out_longer_256"] = int(x["out_len"] > 256)
  x["out_longer_128"] = int(x["out_len"] > 128)
  x["out_longer_64"] = int(x["out_len"] > 64)
  return x

sample_size = 10000
data_stats = retrain_dataset.select(range(sample_size)).map(map_to_length, num_proc=4)

def compute_and_print_stats(x):
  if len(x["input_len"]) == sample_size:
    print(
        "Input Mean: {}, %-Input > 256:{},  %-Input > 128:{}, %-Input > 64:{} Output Mean:{}, %-Output > 256:{}, %-Output > 128:{}, %-Output > 64:{}".format(
            sum(x["input_len"]) / sample_size,
            sum(x["input_longer_256"]) / sample_size,
            sum(x["input_longer_128"]) / sample_size,
            sum(x["input_longer_64"]) / sample_size,   
            sum(x["out_len"]) / sample_size,
            sum(x["out_longer_256"]) / sample_size,
            sum(x["out_longer_128"]) / sample_size,
            sum(x["out_longer_64"]) / sample_size,
        )
    )

output = data_stats.map(
  compute_and_print_stats, 
  batched=True,
  batch_size=-1,
)
```

**Tokenize data**

```python
# # tokenize the examples
def convert_to_features(example_batch):
    input_encodings = tokenizer.batch_encode_plus(example_batch['input'], pad_to_max_length=True, max_length=1)
    target_encodings = tokenizer.batch_encode_plus(example_batch['target'], pad_to_max_length=True, max_length=1)

    encodings = {
        'input_ids': input_encodings['input_ids'], 
        'attention_mask': input_encodings['attention_mask'],
        'labels': target_encodings['input_ids'],
        'decoder_attention_mask': target_encodings['attention_mask']
    }

    return encodings
```

```python
train_data = retrain_dataset.map(convert_to_features, batched=True, remove_columns=retrain_dataset.column_names)
test_data = test_dataset.map(convert_to_features, batched=True, remove_columns=test_dataset.column_names)

columns = ['input_ids', 'attention_mask', 'labels', 'decoder_attention_mask']

train_data.set_format(type='torch', columns=columns)
test_data.set_format(type='torch', columns=columns)
     
```

```python

from datasets import load_metric
rouge = load_metric("rouge")

def compute_metrics(pred):
    labels_ids = pred.label_ids
    pred_ids = pred.predictions

    # all unnecessary tokens are removed
    pred_str = tokenizer.batch_decode(pred_ids, skip_special_tokens=True)
    labels_ids[labels_ids == -100] = tokenizer.pad_token_id
    label_str = tokenizer.batch_decode(labels_ids, skip_special_tokens=True)

    rouge_output = rouge.compute(predictions=pred_str, references=label_str, rouge_types=["rouge2"])["rouge2"].mid

    return {
        "rouge2_precision": round(rouge_output.precision, 4),
        "rouge2_recall": round(rouge_output.recall, 4),
        "rouge2_fmeasure": round(rouge_output.fmeasure, 4),
    }
```

```python

def translate_to_sql(text):
    inputs = tokenizer(text, padding='longest', max_length=64, return_tensors='pt')
    input_ids = inputs.input_ids
    attention_mask = inputs.attention_mask
    output = model.generate(input_ids, attention_mask=attention_mask, max_length=64)

    return tokenizer.decode(output[0], skip_special_tokens=True)
```
